```python
test_data=x_test.join(y_test)

test_data['total_rooms']=np.log(test_data['total_rooms'])+1
test_data['total_bedrooms']=np.log(test_data['total_bedrooms'])+1
test_data['households']=np.log(test_data['households'])+1
test_data['population']=np.log(test_data['population'])+1
```

```python
test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

test_data['household_rooms']=test_data['total_rooms']/test_data['households']
test_data['bedroom_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
```

```python
reg.score(x_test,y_test)
```

A regression score of 0.6799 means that the predictive model you are using has an accuracy or performance of approximately 67.99% on the dataset

Now we head towards a more strong model RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor

forest=RandomForestRegressor()

forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

We can see that it scores 0.8067 which tells us that Random Forest is more stronger than Linear Regression

```python
%%time
from sklearn.model_selection import GridSearchCV
forest=RandomForestRegressor()
param_grid={
    "n_estimators":[3,10,30],
    "max_features":[2,4,6,8],
}

grid_search=GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)

grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

Usually these best estimators results in a better score but sometimes it may score worse or wont change much

```python
best_forest=grid_search.best_estimator_
best_forest.score(x_test,y_test)
```

In our case it scored worst.....it means that the features we tuned are not optimal

```python
%%time
forest=RandomForestRegressor()
param_grid={
    "n_estimators":[100,200,300],
    "min_samples_split":[2,4],
    "max_depth":[None,4,8]
}

grid_search=GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)

grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
best_forest=grid_search.best_estimator_
best_forest.score(x_test,y_test)
```

Now it comes out to be better than before

```python
y_pred = best_forest.predict(x_test)
```

```python
y_pred
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--', color='yellow') 
plt.title('Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.grid(True)
plt.show()
```
